```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from skimage.io import imread
from skimage.color import rgb2gray

img = imread('Grayscaleimg.jpg')
gray_img = rgb2gray(img)

data_matrix = gray_img.reshape(gray_img.shape[0], -1)

pca_5_comp = PCA(n_components=5)
data_pca_5 = pca_5_comp.fit_transform(data_matrix)
reconstructed_5 = pca_5_comp.inverse_transform(data_pca_5)

pca_20_comp = PCA(n_components=20)
data_pca_20 = pca_20_comp.fit_transform(data_matrix)
reconstructed_20 = pca_20_comp.inverse_transform(data_pca_20)

pca_50_comp = PCA(n_components=50)
data_pca_50 = pca_50_comp.fit_transform(data_matrix)
reconstructed_50 = pca_50_comp.inverse_transform(data_pca_50)

pca_100_comp = PCA(n_components=100)
data_pca_100 = pca_100_comp.fit_transform(data_matrix)
reconstructed_100 = pca_100_comp.inverse_transform(data_pca_100)

fig, axs = plt.subplots(1, 5, figsize=(15, 5))

axs[0].imshow(gray_img, cmap='gray')
axs[0].set_title('Original')

axs[1].imshow(reconstructed_5.reshape(gray_img.shape), cmap='gray')
axs[1].set_title('5 Components')

axs[2].imshow(reconstructed_20.reshape(gray_img.shape), cmap='gray')
axs[2].set_title('20 Components')

axs[3].imshow(reconstructed_50.reshape(gray_img.shape), cmap='gray')
axs[3].set_title('50 Components')

axs[4].imshow(reconstructed_100.reshape(gray_img.shape), cmap='gray')
axs[4].set_title('100 Components')

for ax in axs:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
